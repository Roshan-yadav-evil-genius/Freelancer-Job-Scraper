```python
import pandas as pd
from collections import Counter
from matplotlib import pyplot as plt
import seaborn as sns

colors = sns.color_palette("muted", n_colors=20)
```

```python
filepath = "./src/ScrapedData.json"
df=pd.read_json(filepath)
df = df.drop('url', axis=1)
```

```python
df.head()
```

```python
languages = []
top=30
for tech_list in df['Tech']:
    languages.extend(tech_list)
    
language_counts=Counter(languages)
languages=dict(language_counts.most_common(top))
keys=list(languages.keys())
values=list(languages.values())
# make bar horizontal
plt.figure(figsize=(20, 10))
plt.title(f"Top {top} Popular Languages")
sns.barplot(x=values, y=keys, orient='h')
plt.show()
```

```python
# get those jobs which have python and php in there tech

languages = []
popularWith="WordPress"
for tech_list in df[df['Tech'].apply(lambda x: popularWith in x)].Tech:
    languages.extend(tech_list)
    
language_counts=Counter(languages)
del language_counts[popularWith]
languages=dict(language_counts.most_common(20))
keys=list(languages.keys())
values=list(languages.values())

plt.figure(figsize=(20,10))
plt.title("TechStacks Popular With "+popularWith)
sns.barplot(x=values, y=keys, orient='h')
plt.show()
```
